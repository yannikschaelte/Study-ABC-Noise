```python
import pickle
import matplotlib.pyplot as plt

exact_data = pickle.load(open('exact_data.dat', 'rb'))
noisy_data = pickle.load(open('noisy_data.dat', 'rb'))

print(exact_data.keys())
```

```python
plt.plot(noisy_data['growth_curve'], 'x-', label='noisy')
plt.plot(exact_data['growth_curve'], 'x-', label='exact')
plt.legend()

plt.subplots()
plt.plot(noisy_data['extra_cellular_matrix_profile'], '-', label='noisy')
plt.plot(exact_data['extra_cellular_matrix_profile'], '-', label='exact')
plt.legend()

plt.subplots()
plt.plot(noisy_data['proliferation_profile'], '-', label='noisy')
plt.plot(exact_data['proliferation_profile'], '-', label='exact')
plt.legend()
```

```python
fig = plt.figure()

for i, (p, sp) in enumerate([('growth_curve', "Growth curve"), ('extra_cellular_matrix_profile', "ECM profile"),
                             ('proliferation_profile', "Proliferation profile")]):
    ax = fig.add_subplot(3, 1, i+1)
    ax.set_title(sp)
    ax.plot(noisy_data[p], '.-', label='noisy', color='C2')
    ax.plot(exact_data[p], '.-', label='exact', color='grey')
    if i == 0:
        ax.legend()

fig.set_size_inches((3, 8))
fig.tight_layout()
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.8))
plt.savefig("data.svg", format='svg')
```
